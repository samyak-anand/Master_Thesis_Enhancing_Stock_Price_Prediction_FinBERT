```python
# Load Guardian Headlines
print("\nLoading Guardian headlines dataset...")
guardian_headlines = pd.read_csv('/home/samyak/PycharmProjects/Master_Thesis_Enhancing_Stock_Price_Prediction_FinBERT/News_Dataset/guardian_headlines.csv')
print("Shape of Guardian headlines dataset:", guardian_headlines.shape)
print("Columns in Guardian headlines dataset:", guardian_headlines.columns)
```

```python

# Drop duplicates in Guardian Headlines
guardian_headlines = guardian_headlines.drop_duplicates(subset=['Headlines'], keep='first')
guardian_headlines.reset_index(drop=True, inplace=True)
print("Shape after cleaning Guardian headlines dataset:", guardian_headlines.shape)
```

```python

# Load Reuters Headlines
print("\nLoading Reuters headlines dataset...")
reuters_headlines = pd.read_csv('/home/samyak/PycharmProjects/Master_Thesis_Enhancing_Stock_Price_Prediction_FinBERT/News_Dataset/reuters_headlines.csv')
print("Shape of Reuters headlines dataset:", reuters_headlines.shape)
print("Columns in Reuters headlines dataset:", reuters_headlines.columns)
```

```python
# Basic EDA for Reuters Headlines
print("Reuters Headlines Dataset Info:")
print(reuters_headlines.info())
print("Missing Values in Reuters Headlines:")
print(reuters_headlines.isnull().sum())
```

```python
# Drop duplicates in Reuters Headlines
reuters_headlines = reuters_headlines.drop_duplicates(subset=['Headlines', 'Description'], keep='first')
reuters_headlines.reset_index(drop=True, inplace=True)
print("Shape after cleaning Reuters headlines dataset:", reuters_headlines.shape)
```

```python
# Combine All Datasets
print("\nCombining all datasets...")

# Ensure 'Date' is included in all datasets
cnbc_headlines['Time'] = pd.to_datetime(cnbc_headlines['Time'], errors='coerce').dt.date
guardian_headlines['Time'] = pd.to_datetime(guardian_headlines['Time'], errors='coerce').dt.date
reuters_headlines['Time'] = pd.to_datetime(reuters_headlines['Time'], errors='coerce').dt.date
```

```python
combined_news_data = pd.concat(
    [
        cnbc_headlines[['Time', 'Headlines', 'Description']].assign(Source='CNBC'),
        guardian_headlines[['Time', 'Headlines']].assign(Description=None, Source='Guardian'),
        reuters_headlines[['Time', 'Headlines', 'Description']].assign(Source='Reuters')
    ],
    ignore_index=True
)
```
